```
import numpy as np
data = np.load("cell-data.npz")
images = data["images"]
counts = data["counts"]
folds = data["folds"]
X0 = images[folds == 0]
Y0 = counts[folds == 0]
```

```
#Let's pick a single image from the data and show its individual channels
from skimage.color import rgb2hed
import skimage
print('skimage version',skimage.__version__)
import matplotlib.pyplot as plt
I = X0[0]
I_hed =  rgb2hed(I)
plt.imshow(I);plt.title('Original Image');plt.show()
I_h = I_hed[:,:,0]; plt.imshow(I_h,cmap='gray');plt.colorbar();plt.title('H Channel');plt.show()
I_e = I_hed[:,:,1]; plt.imshow(I_e,cmap='gray');plt.colorbar();plt.title('E Channel');plt.show()
I_d = I_hed[:,:,2]; plt.imshow(I_d,cmap='gray');plt.colorbar();plt.title('D Channel');plt.show()
print('Original Image Shape',I.shape)
print('H Channel Shape',I_h.shape)
print('E Channel Shape',I_e.shape)
print('E Channel Shape',I_d.shape)
```
